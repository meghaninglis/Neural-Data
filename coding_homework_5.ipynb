```python
#2 Create a Code Cell that imports relevant modules - specifically, numpy as np, and matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
```

```python
#3 Create a Code Cell that imports these files as numpy arrays called acc_hits and acc_fa, respectively
url = 'https://raw.githubusercontent.com/meghaninglis/Neural-Data/main/acc_fa_hw5.csv'

acc_fa = np.loadtxt(url, delimiter = ',')

url = 'https://raw.githubusercontent.com/meghaninglis/Neural-Data/main/acc_hits_hw5.csv'

acc_hits = np.loadtxt(url, delimiter = ',')
```

```python
#4 Create a Code Cell that prints the shape of each array
print(acc_hits.shape)
print(acc_fa.shape)
```

```python
#6 Create variable called time that has the same length with equally spaced values
time = np.linspace(0,4.0,num=8001)
time
```

```python
#7 Create an expression to determine the sampling rate of the data using the knowledge that you have 8001 samples, and the total length of each sample is 4 seconds

8001/4
```

```python
#8 plot the first trial of acc_hits in one subplot, and the first trial of acc_fa in another subplot. Plot your time variable on the x-axis. Label your axes
fig, ax = plt.subplots(2,2,figsize=(12,6))

ax[0,0].plot(time,acc_hits[0])
ax[0,1].plot(time,acc_fa[0])


plt.xlabel('Time [s]')   
plt.ylabel('LFP')       


```

# #9 Describe what the dominant oscillation in the LFP might be
I think is might be 1.25 becuase there are five distinct peaks over the four second period of time. This does seem to differ between the first trial of acc_hits and the first trial of acc_fa.  There is clearly a different patter of peaks.  There are four distinct peaks over four seconds, leading to a 1.0 dominant oscillation. 

```python
#10 Computing an ERP for acc_hits
plt.plot(time, acc_hits.mean(0))    
plt.xlabel('Time [s]')          
plt.ylabel('LFP voltage')
plt.title('ERP') 

plt.plot([0.25, 0.25], [-0.3,0.3], 'k', lw=2)
plt.hlines(0, time[0], time[-1],'k')
plt.axvline(x=2,color='k')
```

# #11 Interpret your ERP for hits
There does not seem to be much evidence of consistent activity surrounding screen touch.  There is no large spike, but rather small noise-like activities around it. 

```python
#12 Computing an ERP for acc_fa and plot both ERPs in one plot
plt.plot(time, acc_fa.mean(0))       
plt.xlabel('Time [s]')           
plt.ylabel('LFP voltage')
plt.title('ERP') 

plt.plot(time, acc_hits.mean(0), color = 'r')    
plt.xlabel('Time [s]')          
plt.ylabel('LFP voltage')
plt.title('ERP') 

plt.plot([0.25, 0.25], [-0.3,0.3], 'k', lw=2)
plt.hlines(0, time[0], time[-1],'k')
plt.axvline(x=2,color='k')
```

# #9 describe what you think the dominant oscillation in the ACC LFP might be, and what the dominant oscillation in the locus coeruleus (LC) LFP might be

Looking at the hits variable, the dominant oscillation for ACC LFP looks to be around 2.75 because there around 11 distinct peaks over the four second period of time.  Looking at the same hits variable, the dominant oscillation for LC LFP is much lower.  There are very few high peaks on this scale, but if the graph was scaled more for the blue line, there would be a dominant oscillation of about 3.5 because there are about 14 distinct peaks over four seconds.  There is a difference between how the acc and lc hits graphs look.  It appears that the the lc graph is peaking around the same time the acc graph is coming down from its peak, indicating a relationship between the two.  There seems to be a similar relationship for the false alarm graph, with the lc peaking on the downward fall of the acc peak.  The raw LFP does seem to suggest some evidence for relations between peaks of the two. 


```python
#10 compute the coherence for acc_hits and lc_hits

xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits]) 
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])  

dt = time[1] - time[0]
N = lc_hits.shape[1] #this makes the same graph whether you set it as lc or acc



T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  
Syy = scale * (yf * yf.conj()).mean(0)  
Sxy = scale * (xf * yf.conj()).mean(0)  

# Compute the coherence.
cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                     
plt.plot(f, cohr.real, color = 'red')                      
plt.xlim([0, 50])                           
plt.ylim([0, 1])                            
plt.xlabel('Frequency [Hz]')                
plt.ylabel('Coherence')
plt.title('Coherence between two electrodes')
```

# #11 describe your results

There seems to be the highest phase coherence right around one, with a large drop in coherence right after between 3-6 Hz. After 6 Hz, the phase coherence increases sharply again and then begins a somewhat steady decline until the graph ends at 50 Hz. 

```python
#12 repeat step 10, but with acc_fa and lc_fa


xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_fa]) 
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_fa])  

dt = time[1] - time[0]
N = lc_fa.shape[1] #this makes the same graph whether you set it as lc or acc



T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  
Syy = scale * (yf * yf.conj()).mean(0)  
Sxy = scale * (xf * yf.conj()).mean(0)  

# Compute the coherence.
cohrfa = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                     
plt.plot(f, cohrfa.real, color = 'blue')                      
plt.xlim([0, 50])                           
plt.ylim([0, 1])                            
plt.xlabel('Frequency [Hz]')                
plt.ylabel('Coherence')
plt.title('Coherence between two electrodes')
```

```python
#12 putting them on the same plot
plt.plot(f,cohr, color = 'red')
plt.plot(f,cohrfa, color = 'blue')
plt.xlim([0, 50])                           
plt.ylim([0, 1])
plt.xlabel('Frequency [Hz]')                
plt.ylabel('Coherence')
plt.title('Coherence between two electrodes')
```

# #13  intepret your plot from step 12
Both hits and false alarms have their highest coherence right around 1 Hz.  Overall, coherence seems to be higher during hits than false alarms, meaning the two brain areas are talking more during hits than they are during false alarms.
